```python
def convolve(image, gx, gy):
    f_gy = np.flip(gy, 0)
    f_gx = np.flip(gx, 1)
    
    h,w = image.shape
    
    gx = np.zeros((h, w))
    gy = np.zeros((h, w))
    
    for i in range(0, h - 1):
        for j in range(0, w - 1):
            for k in range(0, 3):
                for l in range(0, 3):
                    gx[i][j] += image[i + k - 1][j + l - 1] * f_gx[k][l]
                    gy[i][j] += image[i + k - 1][j + l - 1] * f_gy[k][l]
        
    return np.sqrt(gx * gx + gy * gy) / 255
```

```python
gx = [[-1,0,1],[-2,0,2],[-1,0,1]]
gy = [[-1,-2,-1],[0,0,0],[1,2,1]]
show = convolve(gray, np.array(gx), np.array(gy))

plt.imshow(show)
```

```python
# expected the same as this is a basic sobel filter
```

```python
gx = cv2.Sobel(gray, -1, 1, 0, 3)
gy = cv2.Sobel(gray, -1, 0, 1, 3)

plt.imshow(gx)
plt.show()

plt.imshow(gy)
plt.show()
```

```python
total = cv2.addWeighted(cv2.convertScaleAbs(gx), 0.5, cv2.convertScaleAbs(gy), 0.5, 0)
plt.imshow(total)
```

```python
# not exactly the same as Sobel filter is approximation of derivative
```
